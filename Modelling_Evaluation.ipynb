The database 

```python
data = pd.read_csv('Coinbase_BTCUSD_ob_10_2019_09_01.csv.gz', compression='gzip',error_bad_lines=False)
```

```python
class RevTypeCompares(object):
     def __init__(self, value):
         self.value = value
     def __gt__(self, other):
         return self.value < other[0:len(self.value)]
     def __lt__(self, other):
         return self.value[0:len(self.value)] > other

def find_mid_price(i):
    idx = bisect.bisect_left(list(i['type'].values), RevTypeCompares('a'))
    low = i['price'].values[0]
    high = i['price'].values[idx]

    return (low + high )/2
```

```python
# Get list of prices
prices = []
for date, i in data.groupby("date"):
    mid_price =  find_mid_price(i)
    prices.append(round(mid_price,3))
```

```python
# Set price change between two timestamps as the target variable
price_change = pd.Series(prices).pct_change()[1:]

price_change = price_change.append(pd.Series([0]),ignore_index=True)
price_change[:-1]
```

```python
# Adding target to the other bins shifted backwards 1 step
newdf['price'] = price_change
newdf = newdf[:-1]
target = newdf['price']
newdf.drop('price',axis=1,inplace=True)
newdf
```

Since in the previous steps we noticed the timestamp difference is not consistent and since each timestep is a snapshot of the orderbook the case can be made that they can be investigated as discrete values rather than a continous time-series

```python
# Setting the X and y variables for use in logistic regression
X = newdf
y = target
def Boolean(x):
    if x>0:
        x=True
        return x
    else:
        x=False
        return x
target = target.map(lambda x:int(x>0))
# 0: negative 1: Positive
```

Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
X_train_raw, X_test, y_train_raw, y_test = train_test_split(X, target, test_size=0.2,shuffle=False)
X_train, X_val, y_train, y_val = train_test_split(X_train_raw, y_train_raw,test_size=0.2,shuffle=False)
print("The target variable:")
y_train
```

**Logistic Regression baseline:**

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1e9, solver='lbfgs')
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_val)

df = pd.DataFrame({'validation': y_val, 'linear_predicted': y_pred})
print("The model has returned its predicted values of the validation sample")
df['linear_abs_difference'] = abs(df['validation'] - df['linear_predicted'])
df['linear_abs_difference']
print("The prdiction was wrong {} out of {} entries \nAn accuracy of {}%".format(df['linear_abs_difference'].sum(),df.shape[0],round((1-df['linear_abs_difference'].sum()/df.shape[0])*100,2)))

df
```
